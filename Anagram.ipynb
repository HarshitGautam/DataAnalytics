```python
def test(word):
    return ''.join(sorted(word))
```

```python
def anagram(word):
    return [w for w in wordunique if test(w) == test(word)]
```

```python
anagram('dictionary')
```

```python
anagramDictionary = {}
for word in wordunique:
    anagramDictionary[word] = anagram(word)
```

```python
import collections
```

```python
words_bysig = collections.defaultdict(list)

for word in wordunique:
    words_bysig[test(word)].append(word)
```

```python
words_bysig
```

```python
def anagram_fast(word):
    return words_bysig[test(word)]
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordunique if len(anagram_fast(word)) > 1}
```
