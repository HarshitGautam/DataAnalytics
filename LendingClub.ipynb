# Compute a 95% and 99% confidence interval estimate for the mean interest rate in the training data set. What can you say in a statistical sense about the mean interest rate of loans issued through Lending Club?

### Here is a solution for 95% CI using Statsmodels

```python
import statsmodels.stats.api as sms
sms.DescrStatsW(df.int_rate).tconfint_mean()
```

### Here is a solution for 99% CI using scipy.stats

```python
import scipy.stats as st

st.t.interval(0.99, len(df.int_rate)-1, loc=np.mean(df.int_rate), scale=st.sem(df.int_rate))
```

### Here is a solution for 99% CI using basic Mathematics

Start with looking up the z-value for your desired confidence interval from a look-up table. The confidence interval is then mean +/- z*sigma, where sigma is the estimated standard deviation of your sample mean, given by sigma = s / sqrt(n), where s is the standard deviation computed from your sample data and n is your sample size.



```python
import scipy.stats as st

#zValue = st.norm.ppf(.05)
#zValue

tValue = st.t.ppf(0.01, 3999)
tValue
```

```python
mean = df.int_rate.mean()
mean
```

```python
standardDeviation = df.int_rate.std()
standardDeviation
```

```python
import math
sigma = math.sqrt(standardDeviation/ (df.int_rate.count() - 1))
sigma
```

```python
HCI = mean + (tValue * sigma)
HCI
```

```python
LCI = mean - (tValue * sigma)
LCI
```

For simplicity, you may use debt consolidation, credit card, and home improvement as distinct purposes, and other category. Are the mean interest rates between these purposes significantly different?
```python
F, p = stats.f_oneway(df[df.purpose == 'debt_consolidation']['int_rate'], df[df.purpose == 'credit_card']['int_rate']
                      , df[df.purpose == 'home_improvement']['int_rate'], df[df.purpose == 'debt_consolidation']['int_rate'],
                     df[df.purpose == 'other']['int_rate'])
```

```python
df.term = np.float64(df.term)
df.dti = np.float64(df.dti)
df.inq_last_6mths = np.float64(df.inq_last_6mths)
df.mths_since_last_major_derog = np.float64(df.mths_since_last_major_derog)
df.pub_rec = np.float64(df.pub_rec)
df.open_acc = np.float64(df.open_acc)
df.int_rate = np.float64(df.int_rate)
df.annual_inc = np.float64(df.int_rate)
```

```python
x = []
for i in df['annual_inc']:
    if i != 0.0:
        x.append(math.log(i))
    else:
        x.append(0)
df['log of annual_inc'] = x
y = df['int_rate'].values
X1 = df['log of annual_inc'].values
X1 = sm.add_constant(X1)
model1 = sm.OLS(y,X1)
results1 = model1.fit()
print results1.summary(xname=['Constant','annual_inc'], yname='int_rate')
```

# Multiple Regression with dummy variables

```python
purpose_dummies = pd.get_dummies(df['purpose']).values
homeOwnership_dummies = pd.get_dummies(df['home_ownership']).values
verificationStatus_dummies = pd.get_dummies(df['verification_status']).values
```

```python
purpose_dummies
```

```python
y = df['int_rate']
X = df[['log of annual_inc', 'term', 'dti', 'inq_last_6mths', 'pub_rec', 'open_acc'  ]]
X = sm.add_constant(X)
purpose_cols = list(df['purpose'].unique())
homeOwnership_cols =list(df['home_ownership'].unique())
verificationStatus_cols = list(df['verification_status'].unique())
X = np.concatenate([X, purpose_dummies[:,1:]], axis=1)
X = np.concatenate([X, homeOwnership_dummies[:,1:]], axis=1)
X = np.concatenate([X, verificationStatus_dummies[:,1:]], axis=1)
model2 = sm.OLS(y,X)
results2 = model2.fit()

print results2.summary(xname=['Constant','log of annual_inc', 'term', 'dti', 'inq_last_6mths', 'pub_rec', 'open_acc' ] + purpose_cols[1:] +homeOwnership_cols[1:] + verificationStatus_cols[1:], yname='int_rate')
```
